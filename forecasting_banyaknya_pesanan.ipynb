# Total Transaction for a year

```python
df['createdAt'] = pd.to_datetime(df['createdAt'])
df['date'] = df['createdAt'].dt.date
df['year'] = df['createdAt'].dt.year
df['month'] = df['createdAt'].dt.month
```

```python
def add_value_labels(ax, spacing=5):
    for rect in ax.patches:
        y_value = rect.get_height()
        x_value = rect.get_x() + rect.get_width() / 2
        space = spacing
   
        va = 'bottom'

        if y_value < 0:
            space *= -1
            va = 'top'

        label = "{:.1f}".format(y_value)

        ax.annotate(
            label,                     
            (x_value, y_value),         
            xytext=(0, space),          
            textcoords="offset points", 
            ha='center',                
            va=va)        


plt.figure(figsize=(20,5))
g = sns.countplot(x='month', data=df)
g.set_xticklabels(g.get_xticklabels(),rotation=45)
plt.title('Total transactions for a year ( July2020 - June 2021 )')              
                                        
add_value_labels(g)
```

```python
march = df.loc[df['month']==3]
plt.figure(figsize=(20,5))
m = sns.countplot(x=march['date'], data=df)
m.set_xticklabels(m.get_xticklabels(),rotation=45)
plt.title('Total transactions in March')
add_value_labels(m)
```

```python
des = df.loc[df['month']==12]
plt.figure(figsize=(20,5))
d = sns.countplot(x=des['date'], data=df)
d.set_xticklabels(d.get_xticklabels(),rotation=45)
plt.title('Total transactions in December')
add_value_labels(d)
```

```python
# check prophet version
import fbprophet
# print version number
print('Prophet %s' % fbprophet.__version__)
```

```python
year = df.loc[df['year']==2021]
date2 = year.groupby(by=['date']).count()
trans_by_date = date2['oid'].reset_index(name="y")
trans_by_date
```

```python
trans_by_date['ds'] = pd.to_datetime(trans_by_date['date'])
trans_by_date.drop('date',axis=1)
```

```python
from fbprophet import Prophet
from sklearn.metrics import mean_absolute_error

future = list(trans_by_date.iloc[-30:]['ds'])
train = trans_by_date.drop(trans_by_date.index[-30:])


# define the model
model = Prophet()
# fit the model
model.fit(train)

# define the period for which we want a prediction
future = pd.DataFrame(future)
future.columns = ['ds']
future['ds'] = pd.to_datetime(future['ds'])


# use the model to make a forecast
forecast = model.predict(future)


# calculate MAE between expected and predicted values for december
y_true = trans_by_date['y'][-30:].values
y_pred = forecast['yhat'].values
mae = mean_absolute_error(y_true, y_pred)
print('MAE: %.3f' % mae)

# plot expected vs actual
plt.title("Total transaction")

# pyplot.axis([0, 12, 0, 700000])
plt.plot(y_true, label='Actual')
plt.plot(y_pred, label='Predicted')
plt.legend()
plt.show()
model = Prophet()

# fit the model
model.fit(trans_by_date)
```

```python
from sklearn.metrics import mean_squared_error
from datetime import timedelta

data_to_predict = list()
last_date = max(trans_by_date['ds'])

for i in [7,14,30]:
    data_to_predict.append(last_date + timedelta(days=i))
    
data_to_predict = pd.DataFrame(data_to_predict)
data_to_predict.columns = ['ds']
data_to_predict['ds'] = pd.to_datetime(data_to_predict['ds'])
forecast_real = model.predict(data_to_predict)
forecast_table_trans = forecast_real[['ds', 'yhat']]
forecast_table_trans.rename(columns = {'ds': 'date', 'yhat': 'total_transactions'}, inplace = True)
```

```python
forecast_table_trans.plot(x='date')
plt.title("Total transaction in July")
plt.show()
```
